# My First Jupyter Notebook

This notebook defines functions for `sin(x)` and `cos(x)`, tabulates their values from 0 to 2 using 1000 entries, and prints the first 10 values in columns.

## Sine Function

This function takes a value `x` as input and returns `sin(x)`.

```python
import math

def my_sin(x):
    return math.sin(x)
```

## Cosine Function

This function takes a value `x` as input and returns `cos(x)`.

```python
def my_cos(x):
    return math.cos(x)
```

## Tabulating Values

This cell creates 1000 values of `x` between 0 and 2, then uses the previously defined functions to calculate `sin(x)` and `cos(x)` for each value.

```python
x_values = []
sin_values = []
cos_values = []

start = 0.0
end = 2.0
num_entries = 1000
step = (end - start) / (num_entries - 1)

for i in range(num_entries):
    x = start + i * step
    x_values.append(x)
    sin_values.append(my_sin(x))
    cos_values.append(my_cos(x))
```

## Printing the First 10 Values

This cell uses a `for` loop to print the first 10 values of `x`, `sin(x)`, and `cos(x)` in columns.

```python
print(f"{'x':>10} {'sin(x)':>15} {'cos(x)':>15}")
print("-" * 40)

for i in range(10):
    print(f"{x_values[i]:10.6f} {sin_values[i]:15.6f} {cos_values[i]:15.6f}")
```
