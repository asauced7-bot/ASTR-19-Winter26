# This notebook defines functions for sin(x) and cos(x), tabulates them from 0 to 2 using 1000 entries, and prints the first 10 values.

## import math

def my_sin(x):
    return math.sin(x)

### This function takes a value `x` and returns `sin(x)`.

#### def my_cos(x):
    return math.cos(x)

##### This function takes a value `x` and returns `cos(x)`.

###### x_values = []
sin_values = []
cos_values = []

start = 0.0
end = 2.0
num_entries = 1000
step = (end - start) / (num_entries - 1)

for i in range(num_entries):
    x = start + i * step
    x_values.append(x)
    sin_values.append(my_sin(x))
    cos_values.append(my_cos(x))

This cell creates 1000 values of `x` between 0 and 2, then calculates `sin(x)` and `cos(x)` using the functions defined above.
