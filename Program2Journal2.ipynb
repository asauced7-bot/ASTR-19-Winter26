```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

```python
y_sin = sin_function(x)
y_cos = cos_function(x)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, y_sin)
axes[0].set_title("sin(x)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")

axes[1].plot(x, y_cos)
axes[1].set_title("cos(x)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")

plt.tight_layout()
plt.savefig("sine_cosine_multipanel.pdf")
plt.show()
```
