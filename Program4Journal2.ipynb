```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
random_numbers = rng.exponential(scale=1.0, size=1000)
```

```python
plt.hist(random_numbers, bins=100)
plt.xlabel("Random Number Value")
plt.ylabel("Frequency")
plt.title("Histogram of 1000 Exponential Random Numbers")
plt.savefig("exponential_histogram.pdf")
plt.show()
```
